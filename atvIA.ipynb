```python
fig = plt.figure(figsize=(7, 3))
axes = fig.add_axes([0, 0, 1, 1])
f1 = 'ash'
f2 = 'alcalinity_of_ash'

c = {0: 'r', 1: 'b', 2: 'k'}

for classe in y.unique():
  axes.scatter(X[y==classe][f1], X[y==classe][f2], color=c[classe])
axes.set_xlabel('ash')
axes.set_ylabel('alcalinity_of_ash')
axes.legend(['class_0', 'class_1', 'class_2'])
```
